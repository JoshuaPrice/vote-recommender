```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
summ16 = pd.read_csv('webl16.txt',sep='|',header=None)
```

```python
# WebL Data Dictionary Format
summ16.columns = ['ID','Name','Incumbent','Party Code','Party','Total Receipts','From PACs', 
              'Disbursements','To PACs','Beginning $','End $','From Candidate','Loans From Candidate',
              'Other Loans','Candidate Loan Repayments','Other Loan Repayments','Debt Amount',
              'Individual Contributions','State','District','Special Election','Primary Election',
              'Runoff Election','General Election','General Election %','From Other PACs',
              'From Party','End Date','Refunds to Individuals','Refunds to Committees']
summ16.head()
```

```python
summ16.sort_values('Total Receipts',ascending=False)
```

```python
# Top presidential fundraisers
allPres16 = summ16[summ16['State']=='00'].sort_values('Total Receipts',ascending=False)
pres16 = allPres16[allPres16['Total Receipts'] > 0.0]
```

```python
# Top fundraisers in the Senate
allSenate16 = summ16[(summ16['District']==0.0) & (summ16['State']!='00')].sort_values('Total Receipts',ascending=False)
senate16 = allSenate16[allSenate16['Total Receipts'] > 0.0]
```

```python
# Top fundraisers in the House
allHouse16 = summ16[summ16['District']!=0.0].sort_values('Total Receipts',ascending=False)
house16 = allHouse16[allHouse16['Total Receipts'] > 0.0]
```

```python
# Plot biggest Presidential fundraisers
sns.barplot(pres16.head(20)['Name'], pres16.head(20)['Total Receipts'])
plt.xticks(rotation=90)
```

```python
# Plot biggest Senate fundraisers
sns.barplot(senate16.head(20)['Name'], senate16.head(20)['Total Receipts'])
plt.xticks(rotation=90)
```

```python
# Plot biggest House fundraisers
sns.barplot(house16.head(20)['Name'], house16.head(20)['Total Receipts'])
plt.xticks(rotation=90)
```
